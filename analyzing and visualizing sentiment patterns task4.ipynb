```python
#task4
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r"C:\Users\000GS6744\Downloads\twitter_training.csv.zip"  # Update with the actual file path
df = pd.read_csv(file_path, compression='zip')

# Display the column names
print("Column Names:", df.columns)

# Display the first few rows of the dataset to understand its structure
print("First few rows of the dataset:")
print(df.head())

# Explore the dataset
print(df.info())

# Strip any extra spaces from the column names
df.columns = df.columns.str.strip()

# Renaming columns for better understanding
df.columns = ['ID', 'Entity', 'Sentiment', 'Text']

# Check if the 'Sentiment' column exists
if 'Sentiment' in df.columns:
    # Visualization: Sentiment Distribution
    plt.figure(figsize=(8, 6))
    sns.countplot(x='Sentiment', data=df, palette='Set2')
    plt.title('Sentiment Distribution')
    plt.xlabel('Sentiment')
    plt.ylabel('Count')
    plt.show()
else:
    print("The column 'Sentiment' does not exist in the DataFrame.")
```
